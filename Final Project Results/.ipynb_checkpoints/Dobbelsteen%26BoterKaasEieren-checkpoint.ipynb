# Dobbelsteen

In de opvolgende 3 cells zijn we bezig met het voorspellen van de tegenovergestelde waarde van een dobbelsteen. Hierbij beginnen we simpel door gewoon een single list te vullen met de waardes en labels ( Opdracht 1a), vervolgens veranderen we de shape van deze labels ( Opdracht 1b), op een one-hot encoded way. Hierdoor heb je een 2d array met per value een array die een waarde op 1 heeft en de rest op nul.

Tot slot (Opdracht 2) maken we in de laatste Dobbelsteen cell ook de input 'eyes' one hot encoded. 

Een van de grootste verschillen is de snelheid waarop het model leert. Hoe meer je met one-hot encoded values werkt des te beter het model leert.

```python
# 1 a
import tensorflow as tf

eyes = [1,2,3,4,5,6]
labels = [6,5,4,3,2,1]

#Maak een placeholder om later te vullen met data
X = tf.placeholder(tf.float32, name='X')

#Maak een variabele die door de optimizer veranderd kan worden.
W = tf.Variable(tf.random_normal([1]), name='W')

#Maak nog een variabele die door de optimizer veranderd kan worden.
b = tf.Variable(tf.random_normal([1]), name='b')

#Bereken de kansen voor elke klasse
Y = (tf.multiply(X, W) + b)

#Maak een placeholder om later te vullen met labels
Y_ = tf.placeholder(tf.float32, name='labels')

#Een functie die berekent hoe erg de werkelijke anders is dan de predicted
cross_entropy = tf.reduce_sum(tf.square(Y_ - Y))

#Check of het voorspelde waardes gelijk zijn aan de werkelijke waardes.
is_correct = tf.equal(Y, Y_)
#Bereken de accuracy voor een set waardes
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))

#Definieer met welke methode je wil optimaliseren met de learning rate
optimizer = tf.train.GradientDescentOptimizer(0.01)
train_step = optimizer.minimize(cross_entropy)

#Initieer alle variabelen
init = tf.global_variables_initializer()


#Start een tensorflow sessie
with tf.Session() as sess:
    sess.run(init)
    
    for i in range(1000):
        # load batch of images and correct answers
        batch_X, batch_Y = eyes,labels
        train_data={X: batch_X, Y_: batch_Y}

        # train
        sess.run(train_step, feed_dict=train_data)
        w,c = sess.run([is_correct,cross_entropy], feed_dict=train_data)
#         print(w,c)

    print(sess.run([Y], feed_dict={X:4}))
```

```python
# 1 b
import tensorflow as tf
import numpy as np

eyes = np.array([[1],[2],[3],[4],[5],[6]])
labels = np.eye(6)[[5,4,3,2,1,0]]

X = tf.placeholder(tf.float32, [None,1] ,name='X')
W = tf.Variable(tf.random_normal([1,6]), name='W')
b = tf.Variable(tf.random_normal([6]), name='b')

Y = tf.nn.softmax(tf.multiply(X, W) + b)

Y_ = tf.placeholder(tf.float32, [None,6], name='labels')

cross_entropy = -tf.reduce_sum(Y_ * tf.log(Y))

is_correct = tf.equal(tf.argmax(Y,1), tf.argmax(Y_,1))
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))

optimizer = tf.train.GradientDescentOptimizer(0.02)
train_step = optimizer.minimize(cross_entropy)

init = tf.global_variables_initializer()

prediction = tf.argmax(Y,1)[0] + 1

with tf.Session() as sess:
    sess.run(init)
    
    for i in range(2500):
        # load batch of images and correct answers
        batch_X, batch_Y = eyes,labels
        train_data={X: batch_X, Y_: batch_Y}

        # train
        sess.run(train_step, feed_dict=train_data)
        if i % 100 == 0:
            d,c = sess.run([accuracy,cross_entropy], feed_dict=train_data)
            print(d,c)
            
    print(sess.run([prediction], feed_dict={X:[[4]]}))
```

```python
# 2
import tensorflow as tf
import numpy as np

eyes = np.eye(6)[[0,1,2,3,4,5]]
labels = np.eye(6)[[5,4,3,2,1,0]]

X = tf.placeholder(tf.float32, [None,6] ,name='X')
W = tf.Variable(tf.random_normal([6,6]), name='W')
b = tf.Variable(tf.random_normal([6]), name='b')

Y = tf.nn.softmax(tf.matmul(X, W) + b)

Y_ = tf.placeholder(tf.float32, [None,6], name='labels')

cross_entropy = -tf.reduce_sum(Y_ * tf.log(Y))

is_correct = tf.equal(tf.argmax(Y,1), tf.argmax(Y_,1))
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))

optimizer = tf.train.GradientDescentOptimizer(10)
train_step = optimizer.minimize(cross_entropy)

init = tf.global_variables_initializer()

prediction = tf.argmax(Y,1)

with tf.Session() as sess:
    sess.run(init)
    
    for i in range(10):
        # load batch of images and correct answers
        batch_X, batch_Y = eyes,labels
        train_data={X: batch_X, Y_: batch_Y}

        # train
        sess.run(train_step, feed_dict=train_data)
        if i % 9 == 0:
            y,p,d,c = sess.run([Y,prediction,accuracy,cross_entropy], feed_dict=train_data)
            print(y,p,d,c)
            
    print(sess.run([prediction], feed_dict={X:[[0,0,0,1,0,0]]}))
```

# Boter, Kaas en eieren

Onderstaand hebben wij de tensorflow oplossing voor het voorspellen van een boterkaas en eieren spel. In de eerste cell hebben wij de code om random uitkomsten te genereren staan. 

Voor ons spel hebben we de volgende invultypes bedacht:

<ul>
<li> Niet ingevuld: 0</li>
<li> Speler 1: 1</li>
<li> Speler 2: 2</li>
</ul>


Daarnaast hebben we de labels bedacht:
<ul>
<li> 0: Remise</li>
<li> 1: Speler 1 gewonnen</li>
<li> 2: Speler 2 gewonnen</li>
</ul>



```python
# 3
import numpy as np
from random import shuffle
import random
from functools import reduce

# Geneer nof_games games.
def generateGames(nof_games):
    random.seed(25)
    
    games = []
    labels = []
    for game in range(nof_games):
        newLabel,newGame = generateGame()
        games.append(newGame.flatten())
        labels.append(newLabel)
        
    return np.array(games),labels
        
# Check of er een speler gewonnen heeft, en zo ja, wie?
def getWinner(match):
    winner = winnerByHorizontal(match)
    if haveWinner(winner):
        return winner
    
    winner = winnerByVertical(match)
    if haveWinner(winner):
        return winner
    
    winner = winnerByDiagonal(match)
    if haveWinner(winner):
        return winner
    
    return None
            
# Check of er een speler horizontaal gewonnen heeft
def winnerByHorizontal(match):
    
    def horizontalReduce(row):
        rowMapped = map( lambda v: 3 * 2 if v == 2 else v, row)
        rowSum = reduce( lambda t,v: t + v, rowMapped)
        return rowSum
    
    for row in match:
        rowSum = int(horizontalReduce(row))
        if rowSum == 3:
            return 1
        elif rowSum == 18:
            return 2
            
    return None
  
# Check of er een speler verticaal gewonnen heeft          
def winnerByVertical(match):
    
    def verticalReduce(column):
        columnMapped = map( lambda v: 3 * 2 if v == 2 else v, column)
        columnSum = reduce( lambda t,v: t + v, columnMapped)
        return columnSum
    
    for column in match.T:
        columnSum = int(verticalReduce(column))
        
        if columnSum == 3:
            return 1
        elif columnSum == 18:
            return 2
            
    return None
    
# Check of er een speler diagonaal gewonnen heeft
def winnerByDiagonal(match):
    if (match[0,0] == 1 and match[1,1] == 1 and match[2,2] == 1) or (match[2,0] == 1 and match[1,1] == 1 and match[0,2] == 1):
        return 1
    elif (match[0,0] == 2 and match[1,1] == 2 and match[2,2] == 2) or (match[2,0] == 2 and match[1,1] == 2 and match[0,2] == 2):
        return 2
    else:
        return None
    
# Heeft er iemand gewonnen
def haveWinner(winner):
    return winner != None
    

# Genereer 1 game
def generateGame():
    game = np.zeros((3,3))
    places = [1,2,3,4,5,6,7,8,9]
    
    shuffle(places)
    
    turn = 0
    winner = None
    
    for place in places:
        game[int(place / 3) - 1, place % 3] = turn % 2 + 1
        winner = getWinner(game)
        
        if haveWinner(winner):
            break
        
        turn += 1
            
    winner = 0 if winner == None else winner
    return winner,game
    
games,labels = generateGames(100)

# print(games)
# print(labels)
```

```python
##
# Player 1: O, Player 2: X
##

def displayGame(game, label):
    
    def convertNumberToNoughtsOrCrosses(number):
        if number == 1:
            return 'O'
        elif number == 2:
            return 'X'
        else:
            return ' '
    
    game = [convertNumberToNoughtsOrCrosses(value) for value in game]
    game = np.array(game)
    game = game.reshape(3,3)
    
    stringBuild = ""
    for row in game:
        stringBuild += '| '
        for columnValue in row:
            stringBuild += columnValue + ' '
            
        stringBuild += '|\n'
        
    print(stringBuild)
    
    if label != 0:
        print("Player {} won!".format(convertNumberToNoughtsOrCrosses(label)))
    else:
        print("It's a draw!")
    
gameIndex = 8
displayGame(games[gameIndex], labels[gameIndex])
```

```python
import tensorflow as tf
import numpy as np    

nof_games = 5000
nof_test_games = 500

testGames,testLabels = generateGames(nof_test_games)
testLabels = np.eye(3)[testLabels]
games,labels = generateGames(nof_games)
labels = np.eye(3)[labels]

features = games[0].size
classes = labels[0].size
sndLayerInputNeurons = 46

X = tf.placeholder(tf.float32, [None, features] ,name='X')

W1 = tf.Variable(tf.truncated_normal([features, sndLayerInputNeurons], stddev=0.1), name='W')
b1 = tf.Variable(tf.random_normal([sndLayerInputNeurons]))

Y1 = tf.nn.relu6(tf.matmul(X, W1) + b1) 

# Maak layer 2
W2 = tf.Variable(tf.truncated_normal([sndLayerInputNeurons,classes]))
b2 = tf.Variable(tf.random_normal([classes]), name='b')

# Y = tf.nn.softmax(tf.matmul(Y1, W2) + b2)
# Gebruik layer 2
Ylogits = tf.matmul(Y1, W2) + b2
Y = tf.nn.softmax(Ylogits)

Y_ = tf.placeholder(tf.float32, [None,classes], name='labels')

# cross_entropy = tf.reduce_mean(-tf.reduce_sum(Y_ * tf.log(Y + 1e-10), axis=[1]))#y == [[],[],[]]
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=Ylogits,labels= Y_)
cross_entropy = tf.reduce_mean(cross_entropy)

is_correct = tf.equal(tf.argmax(Y,1), tf.argmax(Y_,1))
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))

optimizer = tf.train.GradientDescentOptimizer(0.045)
train_step = optimizer.minimize(cross_entropy)

prediction = tf.argmax(Y,1)

init = tf.global_variables_initializer()

batch_size = 10
batch_size_fit = int(nof_games / batch_size)

with tf.Session() as sess:
    sess.run(init)
    
    test_data={X: testGames, Y_: testLabels}
    
    for i in range(2000):
        for next_batch in range(batch_size_fit):
            # load batch of images and correct answers
            prevBatchPoint = int(next_batch * batch_size)
            thisBatchPoint = int(next_batch * batch_size + batch_size)

            batch_X, batch_Y = games[prevBatchPoint: thisBatchPoint, :],labels[prevBatchPoint: thisBatchPoint, :]
            
            train_data={X: batch_X, Y_: batch_Y}

            # train
            sess.run(train_step, feed_dict=train_data)
            
        if (i + 1) % 10 == 0:
            p,a,c = sess.run([prediction,accuracy, cross_entropy], feed_dict=test_data)
            print(np.bincount(p),a,c)
            
    print(sess.run([prediction], feed_dict={X:[[ 1 , 2,  2,  2 , 1,  1,  1,  2,  1]]})[0])
```
